```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
```

```python
df=pd.read_csv('cars1.csv')
df.head()
```

```python
sns.pairplot(df,kind='reg')
plt.show()
```

```python
a=pd.DataFrame(df['hp'])
b=pd.DataFrame(df['qsec'])
d2=pd.concat([a,b],axis=1,keys=['hp','qsec'])
d2.head()
```

```python
d2.plot(kind='scatter',x='hp',y='qsec')
```

```python
lm=linear_model.LinearRegression()
model=lm.fit(a,b)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(a,b)
```

```python
xn=([175,101,107,108.109,110,95,120.111,121,105,104])
n=pd.DataFrame(xn)
p=model.predict(n)
p=pd.DataFrame(p)
d3=pd.concat([n,p],axis=1,keys=['hp','qsec'])
```

```python
df.plot(kind='scatter',x='hp',y='qsec')
plt.plot(a,model.predict(a),color='r')
plt.scatter(n,p,color='black')


```
